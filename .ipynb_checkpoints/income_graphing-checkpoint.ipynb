```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading in normalized data & cleaning up
normalized_counts = pd.read_csv('normalized_counts.csv')
del normalized_counts['Unnamed: 0']
normalized_counts.drop(8, inplace=True)
normalized_counts.sort_values('total_norm', ascending=False, inplace=True)
normalized_counts.head()
```

```python
# Reading in income data
income = pd.read_csv("census_data/2016_income.csv", dtype={"Less than 10k":np.float64, "10k to 15k":np.float64,
                                                           "15k to 25k":np.float64, "25k to 35k":np.float64,
                                                           "35k to 50k":np.float64, "50k to 75k":np.float64,
                                                           "75k to 100k":np.float64, "100k to 150k":np.float64,
                                                           "150k to 200k":np.float64, "Over than 200k":np.float64})
income.rename(columns={"Unnamed: 0": "State"}, inplace=True)
income.head()
```

```python
# Getting top 5 and bottom 5
top_5 = normalized_counts.head()
bottom_5 = normalized_counts.tail()
```

```python
# Mergin income data
top_5 = top_5.merge(income)
bottom_5 = bottom_5.merge(income)
```

# Top 5 States

```python
# x & y values
x_axis = np.arange(5)
y_values = list(top_5.columns.values[11:])

colors = []
for value in y_values:
    plt.bar(x_axis, top_5[value], label=value)
    
plt.title("Income")
plt.xlabel("State")
plt.ylabel("Percentage of Population")

x_labels = top_5['State']
x_locations = [value for value in np.arange(5)]
plt.xticks(x_locations, x_labels)

plt.legend(bbox_to_anchor=(1,1), title="Income")

plt.savefig("Images/income_top_5.png", bbox_inches="tight")

plt.show()
```
