```python
state_counts['region'] = ""

for value in us_regions_dict.values():
    print(value)
for index, row in state_counts.iterrows():
    state = row['State Abbr']
    for key,value in us_regions_dict.items():
        if state in value:
            state_counts.set_value(index, 'region', key)
    
```
