```python
file_path = "Resources/national_county.csv"

county_data = pd.read_csv(file_path, dtype={'STATEFP':'str', 'COUNTYFP':'str'})
del county_data['UNKNOWN']
county_data
```

```python
states = ['AS', 'GU', 'MP', 'PR', 'UM', 'VI']

for state in states:
    county_data = county_data.loc[county_data['STATE'] != state]
    
county_data.head()
```

```python
county_data.to_csv("Outputs/county_data.csv", index=False)
```
