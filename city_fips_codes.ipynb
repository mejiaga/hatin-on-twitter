```python
import pandas as pd
import requests as req
import csv
import json
```

```python
file_path = "Resources/national_places.csv"
```

```python
# Lists of data
state = []
state_fp = []
place_fp = []
place_name = []
county = []

with open(file_path, 'r', encoding='utf-8', errors='ignore') as fdata:
    national_places = csv.DictReader(fdata)
    
    for row in national_places:
        state.append(row['STATE'])
        state_fp.append(row['STATEFP'])
        place_fp.append(row['PLACEFP'])
        place_name.append(row['PLACENAME'])
        county.append(row['COUNTY'])
```

```python
national_data = {
    "State": state,
    "State FP": state_fp,
    "City": place_name,
    "City FP": place_fp,
    "County": county
}
```

```python
national_data_df = pd.DataFrame(national_data)
national_data_df
```

```python
national_data_df.to_csv("Outputs/national_data.csv", index=False)
```
