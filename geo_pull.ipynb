```python
from gkeys import gkey
import json
import requests as req
import pandas as pd
```

```python
file = 'random_cities.csv'
```

```python
cities_df = pd.read_csv(file, dtype = {'Zip Code': 'str'})
```

```python
cities_df
```

```python
cities_df['new_lat'] = ""
cities_df['new_lng'] = ""

for index, row in cities_df.iterrows():
    city = row['City Name']
    state = row['State Abbreviation']
    url = "https://maps.googleapis.com/maps/api/geocode/json?address=%s,+%s&key=%s" % (city, state, gkey)
    response = req.get(url).json()
    lat = response['results'][0]['geometry']['location']['lat']
    lng = response['results'][0]['geometry']['location']['lng']
    cities_df.set_value(index, 'new_lat', lat)
    cities_df.set_value(index, 'new_lng', lng)
```
